```python
atc = {}
with open("/Users/Midaroh/Desktop/School/Algorithims and Data Structures/Assignments/Parsed_Sample.txt") as cmp:
    
    for line in cmp:
        chromosome,start_position = line.strip().split('\t')
        
        if chromosome in atc:
            atc[chromosome].append(int(start_position))
        
        else:
            atc[chromosome]=[start_position]
            
atc['ChrM'] = atc['mitochondria']
del atc['mitochondria']
atc['ChrC'] = atc['chloroplast']
del atc['chloroplast']
```

```python
cdict={'Chr1': 0,
       'Chr2': 0,
       'Chr3': 0,
       'Chr4': 0,
       'Chr5': 0,
       'ChrM': 0,
       'ChrC': 0
      }

rdict={'Chr1': [],
       'Chr2': [],
       'Chr3': [],
       'Chr4': [],
       'Chr5': [],
       'ChrM': [],
       'ChrC': []
      }

with open("/Users/Midaroh/Desktop/School/Algorithims and Data Structures/Assignments/Parsed_Arabidopsis.txt",'r') as cpr:
    
    for line in cpr:
        data = line.split('\t')

        if data[0] in list(atc.keys()):
            tmpk=data[0]
            rdict[tmpk] += [[int(data[3]),int(data[4])]]
```

```python
#UNSORTED run
import time
start=time.process_time()

for key in atc:
    t=key
    tmpatc=atc.get(t)
    tmprange=rdict.get(t)
    tmpc=0

    if t in list(rdict.keys()):
        for x in tmpatc:
            for y in tmprange:
                if ([y[0],y[1]] != [0,0]):
                    if (int(x)>=int(y[0]) and int(x)<=int(y[1])):
                        tmpc+=1  
                        cdict[t] = tmpc
                        [y[0],y[1]]=[0,0]        

print("Time to complete unsorted run = "+str(time.process_time() - start)+" seconds")
print(cdict)  
```

```python
#Sorting function using merge sort
def mergeSort(grp):
    if len(grp)>1:
        #assign value to the middle of the list
        mid = len(grp)//2
        #split the list in 2, L being the left and R being the right
        L = grp[:mid]
        R = grp[mid:]
        #cal the function to split the Left and Right side indefinetly until only 1 element remains in each split
        mergeSort(L)
        mergeSort(R)
        #creat values to act as trackers
        x = y = z = 0
        #use temporary values to hold and sort value while resorting
        while x < len(L) and y < len(R):
            if int(L[x]) < int(R[y]):
                grp[z] = L[x]
                x += 1
            else:
                grp[z] = R[y]
                y += 1
            z += 1
        #Make sure there are no values left in the splits
        while x < len(L):
            grp[z] = L[x]
            x += 1
            z += 1
        while y < len(R):
            grp[z] = R[y]
            y += 1
            z += 1
```

```python
#sort atc
for t in atc:
    mergeSort(atc.get(t))
```

```python
cdict={'Chr1': 0,
       'Chr2': 0,
       'Chr3': 0,
       'Chr4': 0,
       'Chr5': 0,
       'ChrM': 0,
       'ChrC': 0
      }

rdict={'Chr1': [],
       'Chr2': [],
       'Chr3': [],
       'Chr4': [],
       'Chr5': [],
       'ChrM': [],
       'ChrC': []
      }

with open("/Users/Midaroh/Desktop/School/Algorithims and Data Structures/Assignments/Parsed_Arabidopsis.txt",'r') as cpr:

    for line in cpr:
        data = line.split('\t')
        
        if data[0] in list(atc.keys()):
            tmpk=data[0]
            
            if len(rdict[tmpk])==0:
                rdict[tmpk] += [[int(data[3]),int(data[4])]]
            else:
                rc=-1
                for z in rdict[tmpk]:
                    rc+=1
                    
                    if (int(data[3])==z[0] and int(data[4])==z[1]):
                        break
                        
                    if int(data[3])>z[0]:
                        if rc+1==len(rdict[tmpk]):
                            rdict[tmpk] += [[int(data[3]),int(data[4])]]
                        continue
                        
                    if int(data[3])<z[0]:
                        rdict[tmpk].insert((rc),[int(data[3]),int(data[4])])
                        break
```

```python
#SORTED run
import time
start=time.process_time()

for key in atc:
    t=key
    tmpatc=atc.get(t)
    tmprange=rdict.get(t)
    tmpc=0
    
    if t in list(rdict.keys()):
        for x in tmpatc:
            for y in tmprange:
                if ([y[0],y[1]] != [0,0]):
                    if (int(x)>=int(y[0]) and int(x)<=int(y[1])):
                        tmpc+=1  
                        cdict[t] = tmpc
                        [y[0],y[1]]=[0,0]
                    elif int(x)>int(y[1]):
                        continue

print("Time to complete sorted run = "+str(time.process_time() - start)+" seconds")
print(cdict)
```
